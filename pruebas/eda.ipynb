```python
import psycopg2
import matplotlib.pyplot as plt
import seaborn as sns

def conectar_bd():
    conn = psycopg2.connect(
        host="localhost",
        database="SEPHORA",
        user="postgres",
        password="admin"
    )
    return conn, conn.cursor()

def ejecutar_query(query):
    conn, cur = conectar_bd()
    cur.execute(query)
    cols = [desc[0] for desc in cur.description]
    datos = cur.fetchall()
    cur.close()
    conn.close()
    return pd.DataFrame(datos, columns=cols)
```

```python
query_categoria = """
WITH conteo_categoria AS (
    SELECT 
        c.nombre_categoria,
        m.nombre_marca,
        COUNT(*) AS num_productos
    FROM productos p
    JOIN categorias c ON p.id_categoria = c.id_categoria
    JOIN marcas m ON p.id_marca = m.id_marca
    GROUP BY c.nombre_categoria, m.nombre_marca
),
ranking_categoria AS (
    SELECT *,
           RANK() OVER (PARTITION BY nombre_categoria ORDER BY num_productos DESC) AS pos
    FROM conteo_categoria
)
SELECT nombre_categoria, nombre_marca, num_productos
FROM ranking_categoria
WHERE pos = 1
ORDER BY nombre_categoria;
"""
df_cat = ejecutar_query(query_categoria)
df_cat
```

```python
query_subcategoria = """WITH conteo_subcategoria AS (
    SELECT 
        s.nombre_subcategoria,
        m.nombre_marca,
        COUNT(*) AS num_productos
    FROM productos p
    JOIN subcategorias s ON p.id_subcategoria = s.id_subcategoria
    JOIN marcas m ON p.id_marca = m.id_marca
    GROUP BY s.nombre_subcategoria, m.nombre_marca
),
ranking_subcategoria AS (
    SELECT *,
           RANK() OVER (PARTITION BY nombre_subcategoria ORDER BY num_productos DESC) AS pos
    FROM conteo_subcategoria
)
SELECT nombre_subcategoria, nombre_marca, num_productos
FROM ranking_subcategoria
WHERE pos = 1
ORDER BY nombre_subcategoria;
"""
df_subcat = ejecutar_query(query_subcategoria)
df_subcat
```

```python
# Exploración y Análisis de Datos para el TFM - Sephora Collection

import pandas as pd
import psycopg2
import matplotlib.pyplot as plt
import seaborn as sns

# Conexión a la base de datos
def conectar_bd():
    conn = psycopg2.connect(
        host="localhost",
        database="SEPHORA",
        user="postgres",
        password="admin"
    )
    return conn, conn.cursor()

# Ejecutar consulta SQL y obtener DataFrame
def ejecutar_query(query):
    conn, cur = conectar_bd()
    cur.execute(query)
    cols = [desc[0] for desc in cur.description]
    datos = cur.fetchall()
    cur.close()
    conn.close()
    return pd.DataFrame(datos, columns=cols)

### 1. Análisis General ###

# Productos por Categoría y Subcategoría
query_cat_subcat = """
SELECT c.nombre_categoria, sc.nombre_subcategoria, COUNT(*) AS num_productos
FROM productos p
JOIN categorias c ON p.id_categoria = c.id_categoria
JOIN subcategorias sc ON p.id_subcategoria = sc.id_subcategoria
GROUP BY c.nombre_categoria, sc.nombre_subcategoria
ORDER BY num_productos DESC;
"""
df_cat_subcat = ejecutar_query(query_cat_subcat)

plt.figure(figsize=(12,6))
sns.barplot(data=df_cat_subcat.head(10), x="num_productos", y="nombre_subcategoria", hue="nombre_categoria")
plt.title("Top 10 Subcategorías con más productos")
plt.tight_layout()
plt.show()

# Productos por Marca
query_marcas = """
SELECT m.nombre_marca, COUNT(*) AS num_productos
FROM productos p
JOIN marcas m ON p.id_marca = m.id_marca
GROUP BY m.nombre_marca
ORDER BY num_productos DESC;
"""
df_marcas = ejecutar_query(query_marcas)

plt.figure(figsize=(12,6))
sns.barplot(data=df_marcas, y="nombre_marca", x="num_productos")
plt.title("Top 15 Marcas con más productos")
plt.tight_layout()
plt.show()

# Distribución de Precios
query_precios = """
SELECT precio FROM historico WHERE precio IS NOT NULL;
"""
df_precios = ejecutar_query(query_precios)

plt.figure(figsize=(10,5))
sns.histplot(df_precios['precio'], bins=30, kde=True)
plt.title("Distribución de Precios")
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.tight_layout()
plt.show()

# Valoración promedio por subcategoría
query_val_subcat = """
SELECT sc.nombre_subcategoria, AVG(h.valoracion) AS media_valoracion, SUM(h.numero_valoraciones) AS total_valoraciones
FROM historico h
JOIN productos p ON h.id_producto = p.id_producto
JOIN subcategorias sc ON p.id_subcategoria = sc.id_subcategoria
WHERE h.valoracion IS NOT NULL
GROUP BY sc.nombre_subcategoria
HAVING SUM(h.numero_valoraciones) > 50
ORDER BY media_valoracion DESC;
"""
df_val_subcat = ejecutar_query(query_val_subcat)

plt.figure(figsize=(12,6))
sns.barplot(data=df_val_subcat.head(10), x="media_valoracion", y="nombre_subcategoria")
plt.title("Subcategorías mejor valoradas (con >50 valoraciones)")
plt.tight_layout()
plt.show()

### 2. Oportunidades para Sephora Collection ###

# Subcategorías donde Sephora Collection NO está
query_no_sephora = """
SELECT DISTINCT sc.nombre_subcategoria
FROM productos p
JOIN marcas m ON p.id_marca = m.id_marca
JOIN subcategorias sc ON p.id_subcategoria = sc.id_subcategoria
WHERE m.nombre_marca = 'Sephora Collection';
"""
df_sephora_subcats = ejecutar_query(query_no_sephora)
subcats_sephora = set(df_sephora_subcats['nombre_subcategoria'])

subcats_all = set(df_val_subcat['nombre_subcategoria'])
oportunidades = subcats_all - subcats_sephora

print("\nSubcategorías con buena valoración sin presencia de Sephora Collection:")
for subcat in df_val_subcat['nombre_subcategoria']:
    if subcat in oportunidades:
        print("-", subcat)

### 3. Análisis de Características del Producto Estrella ###

# Características de productos top (ej. Base de maquillaje con valoración > 4.5)
def analizar_caracteristica(tabla, campo, subcat):
    query = f"""
    SELECT {tabla}.{campo}, COUNT(*) AS frecuencia
    FROM productos p
    JOIN producto_{tabla} pf ON p.id_producto = pf.id_producto
    JOIN {tabla} ON pf.id_{tabla} = {tabla}.id_{tabla}
    JOIN historico h ON p.id_producto = h.id_producto
    JOIN subcategorias sc ON p.id_subcategoria = sc.id_subcategoria
    WHERE h.valoracion > 4.5 AND sc.nombre_subcategoria = '{subcat}'
    GROUP BY {tabla}.{campo}
    ORDER BY frecuencia DESC
    LIMIT 10;
    """
    df = ejecutar_query(query)
    print(f"\n{tabla.capitalize()} más comunes en productos top de '{subcat}':")
    print(df)

# Llamar para una subcategoría concreta
subcat_objetivo = 'Base de maquillaje'
for tabla, campo in [
    ("formulaciones", "nombre_formulacion"),
    ("formatos", "nombre_formato"),
    ("responsabilidades", "nombre_responsabilidad"),
    ("efectos_sombra", "nombre_efecto"),
    ("efectos_labios", "nombre_efecto"),
    ("efectos_mascara", "nombre_efecto"),
    ("tipos_piel", "nombre_tipo_piel"),
    ("coberturas", "nombre_cobertura"),
    ("acabados", "nombre_acabado"),
    ("texturas", "nombre_textura")
]:
    analizar_caracteristica(tabla, campo, subcat_objetivo)

### 4. Precio Promedio por Marca en Subcategoría Objetivo ###
query_precio_marca = f"""
SELECT m.nombre_marca, AVG(h.precio) AS precio_medio
FROM historico h
JOIN productos p ON h.id_producto = p.id_producto
JOIN marcas m ON p.id_marca = m.id_marca
JOIN subcategorias sc ON p.id_subcategoria = sc.id_subcategoria
WHERE sc.nombre_subcategoria = '{subcat_objetivo}' AND h.precio IS NOT NULL
GROUP BY m.nombre_marca
ORDER BY precio_medio DESC
LIMIT 15;
"""
df_precio_marca = ejecutar_query(query_precio_marca)

plt.figure(figsize=(12,6))
sns.barplot(data=df_precio_marca, y="nombre_marca", x="precio_medio")
plt.title(f"Precio medio por marca en '{subcat_objetivo}'")
plt.tight_layout()
plt.show()

# Puedes guardar los DataFrames si quieres exportarlos para Excel o PDF
# df_val_subcat.to_csv("subcategorias_valoracion.csv", index=False)
# df_precio_marca.to_csv("precio_por_marca.csv", index=False)  

print("\n--- Exploración finalizada ---")
```
